## Q1.
Bar plots are used to show a numerical comparison across different categories in data visualization. In the following data set, calculate the average alcohol amounts (beer, spirit, wine, and liters) for each continent and compare these values in a bar plot.

```
# Read in the drinks data as a Pandas DataFrame
url = 'https://gist.githubusercontent.com/aegorenkov/f7f0f8d6e9cc17ca9774572e224e2ad1/raw/9429430ee053bc796d694ed7e88b333a54f3e287/drinks.csv'
drink_cols = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent']
drinks = pd.read_csv(url, header=0, na_filter=False)
drinks.head()
```

```
# Answer:

drinks.groupby('continent').mean().plot(kind='bar')
plt.title("The average alcohol amounts for each continents bar plot")
```

## Q2.


Scatter plots are used to display the relationship between two numerical variables in data visualization. Seaborn library in Python is quite helpful in generating such scatter plots https://seaborn.pydata.org/generated/seaborn.scatterplot.html. Using the following mpg dataset from Seaborn's data repository, create a scatter plot of miles per gallon against against horsepower. In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. Also, transparency of a plot is managed by the parameter alpha between 0 an 1. Use these parameters in your scatter plot to enrich your display.

```
import seaborn as sns
sns.set(style="white")

# Load the example mpg dataset
mpg = sns.load_dataset("mpg")
mpg.head()
```

```
# Answer:
x ,y = mpg['mpg'], mpg['horsepower']
color = mpg['origin']
sns.scatterplot(x, y, alpha = 0.5, hue = color)
```

Once you have a cleaned dataset, the next step is Exploratory Data Analysis (EDA). It is a process of exploring patterns, relationship, or anomalies to guide us in our next steps in our data analysis. One of the important tools that we utilize in EDA is the pair plots - a scatterplot matrix. Pair plots display us the distribution of a single variable along with pairwise relationships between variables in our data set. Using the paiplot function in Seaborn library (https://seaborn.pydata.org/generated/seaborn.pairplot.html), explore the following socioeconomic data set obtained from Gapminder (https://www.gapminder.org/data/).
